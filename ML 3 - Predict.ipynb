```python
# Import necessary library for encoding
from sklearn.preprocessing import LabelEncoder

# Identify categorical features in your dataframe
categorical_features = df_predecir.select_dtypes(include=['object']).columns

# Create a LabelEncoder for each categorical feature
for feature in categorical_features:
    # Create and fit the encoder
    encoder = LabelEncoder()
    encoder.fit(df_predecir[feature])
    # Transform the feature in the dataframe
    df_predecir[feature] = encoder.transform(df_predecir[feature])

# Now you can try predicting again:
predicts = modelo_cargado.predict(df_predecir)
predicts_proba = modelo_cargado.predict_proba(df_predecir)
```

# Guarda predicciones

* Guardar las predicciones en data path. Cada fila debe estar etiquetada con el ID.

```python

# Create a DataFrame with customer IDs and predictions
predicciones = pd.DataFrame({'customerID': df_predecir.index, 'Churn_prediction': predicts})

# Save predictions to CSV
predicciones.to_csv('/content/drive/MyDrive/Nuclio/2º entregable/predicciones_churn.csv', index=False)
```
